```python
# GeoPoint 거리 계산 테스트
from math import pi

import app.geo as geo

print("=== GeoPoint 거리 계산 테스트 ===")

# 서울과 부산의 좌표 (대략적)
seoul = geo.GeoPoint.from_deg(37.5665, 126.9780)
busan = geo.GeoPoint.from_deg(35.1796, 129.0756)

print(f"서울: {seoul}")
print(f"부산: {busan}")

# 내부 값 확인 (라디안으로 저장되어야 함)
print("\n=== 내부 저장값 (라디안) ===")
print(f"서울 위도 내부값: {float(seoul.latitude)}")
print(f"서울 경도 내부값: {float(seoul.longitude)}")
print(f"부산 위도 내부값: {float(busan.latitude)}")
print(f"부산 경도 내부값: {float(busan.longitude)}")

# 예상 라디안 값과 비교
print("\n=== 예상 라디안 값과 비교 ===")
print(f"서울 위도 37.5665° → 라디안: {37.5665 * pi / 180}")
print(f"서울 경도 126.9780° → 라디안: {126.9780 * pi / 180}")

# 거리 계산
try:
    distance = seoul.distance_to(busan)
    print("\n=== 거리 계산 결과 ===")
    print(f"서울-부산 거리: {distance}")
    print(f"거리 (km): {float(distance) / 1000:.2f} km")
    print("실제 서울-부산 거리는 약 325km 정도입니다.")
except Exception as e:
    print(f"거리 계산 오류: {e}")
    print(f"오류 타입: {type(e)}")

# 가까운 두 점으로도 테스트
point1 = geo.GeoPoint.from_deg(37.5665, 126.9780)  # 서울
point2 = geo.GeoPoint.from_deg(37.5665, 126.9790)  # 서울에서 경도만 0.001도 차이

try:
    short_distance = point1.distance_to(point2)
    print("\n=== 짧은 거리 테스트 ===")
    print(f"0.001도 경도 차이 거리: {short_distance}")
    print(f"거리 (m): {float(short_distance):.2f} m")
except Exception as e:
    print(f"짧은 거리 계산 오류: {e}")
```

```python
# GeoPoint forward 메서드 테스트
from math import pi

import app.geo as geo
import app.unit as unit

print("=== GeoPoint forward 메서드 테스트 ===")

# 서울 시청 좌표
seoul = geo.GeoPoint.from_deg(37.5665, 126.9780)
print(f"시작점 (서울): {seoul}")

# 테스트 1: 동쪽으로 1km 이동
print("\n=== 테스트 1: 동쪽으로 1km 이동 ===")
east_bearing = unit.Degree(90)  # 동쪽
distance_1km = unit.Meter(1000)  # 1km

try:
    point_east = seoul.forward(east_bearing, distance_1km)
    print(f"동쪽 1km 이동 후: {point_east}")
    print(f"위도 변화: {float(point_east.latitude) - float(seoul.latitude):.8f} 라디안")
    print(f"경도 변화: {float(point_east.longitude) - float(seoul.longitude):.8f} 라디안")

    # 역방향 거리 계산으로 검증
    back_distance = seoul.distance_to(point_east)
    print(f"역방향 거리 계산: {float(back_distance):.2f}m (1000m와 비교)")

except Exception as e:
    print(f"동쪽 이동 테스트 오류: {e}")
    import traceback

    traceback.print_exc()

# 테스트 2: 북쪽으로 10km 이동
print("\n=== 테스트 2: 북쪽으로 10km 이동 ===")
north_bearing = unit.Degree(0)  # 북쪽
distance_10km = unit.Meter(10000)  # 10km

try:
    point_north = seoul.forward(north_bearing, distance_10km)
    print(f"북쪽 10km 이동 후: {point_north}")
    print(f"위도 변화: {float(point_north.latitude) - float(seoul.latitude):.8f} 라디안")
    print(f"경도 변화: {float(point_north.longitude) - float(seoul.longitude):.8f} 라디안")

    # 역방향 거리 계산으로 검증
    back_distance = seoul.distance_to(point_north)
    print(f"역방향 거리 계산: {float(back_distance):.2f}m (10000m와 비교)")

except Exception as e:
    print(f"북쪽 이동 테스트 오류: {e}")
    import traceback

    traceback.print_exc()

# 테스트 3: 남서쪽으로 5km 이동
print("\n=== 테스트 3: 남서쪽으로 5km 이동 ===")
southwest_bearing = unit.Degree(225)  # 남서쪽 (180 + 45)
distance_5km = unit.Meter(5000)  # 5km

try:
    point_sw = seoul.forward(southwest_bearing, distance_5km)
    print(f"남서쪽 5km 이동 후: {point_sw}")
    print(f"위도 변화: {float(point_sw.latitude) - float(seoul.latitude):.8f} 라디안")
    print(f"경도 변화: {float(point_sw.longitude) - float(seoul.longitude):.8f} 라디안")

    # 역방향 거리 계산으로 검증
    back_distance = seoul.distance_to(point_sw)
    print(f"역방향 거리 계산: {float(back_distance):.2f}m (5000m와 비교)")

except Exception as e:
    print(f"남서쪽 이동 테스트 오류: {e}")
    import traceback

    traceback.print_exc()

# 테스트 4: 연속 이동 테스트 (서울 → 동쪽 1km → 북쪽 1km)
print("\n=== 테스트 4: 연속 이동 테스트 ===")
try:
    # 서울에서 동쪽으로 1km
    step1 = seoul.forward(unit.Degree(90), unit.Meter(1000))
    print(f"1단계 (동쪽 1km): {step1}")

    # 그 지점에서 북쪽으로 1km
    step2 = step1.forward(unit.Degree(0), unit.Meter(1000))
    print(f"2단계 (북쪽 1km): {step2}")

    # 서울에서 최종 지점까지의 직선 거리
    final_distance = seoul.distance_to(step2)
    print(f"서울에서 최종 지점까지 직선 거리: {float(final_distance):.2f}m")
    print(f"이론적 거리 (피타고라스): {(1000**2 + 1000**2) ** 0.5:.2f}m")

except Exception as e:
    print(f"연속 이동 테스트 오류: {e}")
    import traceback

    traceback.print_exc()

print("\n=== forward 메서드 테스트 완료 ===")
```

```python
# Degree 나눗셈 문제 분석
from math import pi

import app.unit as unit

print("=== Degree 나눗셈 문제 분석 ===")

degree_90 = unit.Degree(90.0)
degree_45 = unit.Degree(45.0)

print(f"degree_90: {degree_90}")
print(f"degree_45: {degree_45}")

# 내부값 (라디안) 확인
print("\n=== 내부값 (라디안) ===")
print(f"degree_90 내부값: {float(degree_90):.10f} 라디안")
print(f"degree_45 내부값: {float(degree_45):.10f} 라디안")
print(f"90도 라디안: {90 * pi / 180:.10f}")
print(f"45도 라디안: {45 * pi / 180:.10f}")

# 나눗셈 계산
print("\n=== 나눗셈 분석 ===")
result = degree_90 / degree_45
print(f"90° / 45° = {result}")
print(f"결과 내부값: {float(result):.10f}")

# 예상 결과
expected_internal = float(degree_90) / float(degree_45)
print(f"예상 내부값 (라디안/라디안): {expected_internal:.10f}")

# 단위 변환으로 표시값 계산
expected_display = expected_internal * 180 / pi
print(f"예상 표시값 (도): {expected_display:.10f}")

print("\n=== 문제 발견 ===")
print(f"실제 결과: {float(result):.10f}")
print(f"예상 결과: {expected_internal:.10f} (= 2.0)")
print(f"차이: {abs(float(result) - expected_internal):.10f}")

# 다른 계산들도 확인
print("\n=== 다른 연산 확인 ===")
add_result = degree_90 + degree_45
sub_result = degree_90 - degree_45
mul_result = degree_90 * degree_45

print(f"90° + 45° = {add_result} (내부: {float(add_result):.6f})")
print(f"90° - 45° = {sub_result} (내부: {float(sub_result):.6f})")
print(f"90° * 45° = {mul_result} (내부: {float(mul_result):.6f})")

# 예상값과 비교
print("\n=== 예상값과 비교 ===")
print(f"덧셈 예상: {(90 + 45) * pi / 180:.6f} 라디안 = 135°")
print(f"뺄셈 예상: {(90 - 45) * pi / 180:.6f} 라디안 = 45°")
print("곱셈 예상: ??? (단위가 °²이 되어야 함)")
print(f"나눗셈 예상: {90 / 45:.6f} = 2.0 (무차원)")
```
