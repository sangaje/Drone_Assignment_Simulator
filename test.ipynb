```python
# GeoPoint 거리 계산 테스트
from math import pi

import dronesim.geo as geo

print("=== GeoPoint 거리 계산 테스트 ===")

# 서울과 부산의 좌표 (대략적)
seoul = geo.GeoPoint.from_deg(37.5665, 126.9780)
busan = geo.GeoPoint.from_deg(35.1796, 129.0756)

print(f"서울: {seoul}")
print(f"부산: {busan}")

# 내부 값 확인 (라디안으로 저장되어야 함)
print("\n=== 내부 저장값 (라디안) ===")
print(f"서울 위도 내부값: {float(seoul.latitude)}")
print(f"서울 경도 내부값: {float(seoul.longitude)}")
print(f"부산 위도 내부값: {float(busan.latitude)}")
print(f"부산 경도 내부값: {float(busan.longitude)}")

# 예상 라디안 값과 비교
print("\n=== 예상 라디안 값과 비교 ===")
print(f"서울 위도 37.5665° → 라디안: {37.5665 * pi / 180}")
print(f"서울 경도 126.9780° → 라디안: {126.9780 * pi / 180}")

# 거리 계산
try:
    distance = seoul.distance_to(busan)
    print("\n=== 거리 계산 결과 ===")
    print(f"서울-부산 거리: {distance}")
    print(f"거리 (km): {float(distance) / 1000:.2f} km")
    print("실제 서울-부산 거리는 약 325km 정도입니다.")
except Exception as e:
    print(f"거리 계산 오류: {e}")
    print(f"오류 타입: {type(e)}")

# 가까운 두 점으로도 테스트
point1 = geo.GeoPoint.from_deg(37.5665, 126.9780)  # 서울
point2 = geo.GeoPoint.from_deg(37.5665, 126.9790)  # 서울에서 경도만 0.001도 차이

try:
    short_distance = point1.distance_to(point2)
    print("\n=== 짧은 거리 테스트 ===")
    print(f"0.001도 경도 차이 거리: {short_distance}")
    print(f"거리 (m): {float(short_distance):.2f} m")
except Exception as e:
    print(f"짧은 거리 계산 오류: {e}")
```

```python
# GeoPoint forward 메서드 테스트
from math import pi

import dronesim.geo as geo
import dronesim.unit as unit

print("=== GeoPoint forward 메서드 테스트 ===")

# 서울 시청 좌표
seoul = geo.GeoPoint.from_deg(37.5665, 126.9780)
print(f"시작점 (서울): {seoul}")

# 테스트 1: 동쪽으로 1km 이동
print("\n=== 테스트 1: 동쪽으로 1km 이동 ===")
east_bearing = unit.Degree(90)  # 동쪽
distance_1km = unit.Meter(1000)  # 1km

try:
    point_east = seoul.forward(east_bearing, distance_1km)
    print(f"동쪽 1km 이동 후: {point_east}")
    print(f"위도 변화: {float(point_east.latitude) - float(seoul.latitude):.8f} 라디안")
    print(f"경도 변화: {float(point_east.longitude) - float(seoul.longitude):.8f} 라디안")

    # 역방향 거리 계산으로 검증
    back_distance = seoul.distance_to(point_east)
    print(f"역방향 거리 계산: {float(back_distance):.2f}m (1000m와 비교)")

except Exception as e:
    print(f"동쪽 이동 테스트 오류: {e}")
    import traceback

    traceback.print_exc()

# 테스트 2: 북쪽으로 10km 이동
print("\n=== 테스트 2: 북쪽으로 10km 이동 ===")
north_bearing = unit.Degree(0)  # 북쪽
distance_10km = unit.Meter(10000)  # 10km

try:
    point_north = seoul.forward(north_bearing, distance_10km)
    print(f"북쪽 10km 이동 후: {point_north}")
    print(f"위도 변화: {float(point_north.latitude) - float(seoul.latitude):.8f} 라디안")
    print(f"경도 변화: {float(point_north.longitude) - float(seoul.longitude):.8f} 라디안")

    # 역방향 거리 계산으로 검증
    back_distance = seoul.distance_to(point_north)
    print(f"역방향 거리 계산: {float(back_distance):.2f}m (10000m와 비교)")

except Exception as e:
    print(f"북쪽 이동 테스트 오류: {e}")
    import traceback

    traceback.print_exc()

# 테스트 3: 남서쪽으로 5km 이동
print("\n=== 테스트 3: 남서쪽으로 5km 이동 ===")
southwest_bearing = unit.Degree(225)  # 남서쪽 (180 + 45)
distance_5km = unit.Meter(5000)  # 5km

try:
    point_sw = seoul.forward(southwest_bearing, distance_5km)
    print(f"남서쪽 5km 이동 후: {point_sw}")
    print(f"위도 변화: {float(point_sw.latitude) - float(seoul.latitude):.8f} 라디안")
    print(f"경도 변화: {float(point_sw.longitude) - float(seoul.longitude):.8f} 라디안")

    # 역방향 거리 계산으로 검증
    back_distance = seoul.distance_to(point_sw)
    print(f"역방향 거리 계산: {float(back_distance):.2f}m (5000m와 비교)")

except Exception as e:
    print(f"남서쪽 이동 테스트 오류: {e}")
    import traceback

    traceback.print_exc()

# 테스트 4: 연속 이동 테스트 (서울 → 동쪽 1km → 북쪽 1km)
print("\n=== 테스트 4: 연속 이동 테스트 ===")
try:
    # 서울에서 동쪽으로 1km
    step1 = seoul.forward(unit.Degree(90), unit.Meter(1000))
    print(f"1단계 (동쪽 1km): {step1}")

    # 그 지점에서 북쪽으로 1km
    step2 = step1.forward(unit.Degree(0), unit.Meter(1000))
    print(f"2단계 (북쪽 1km): {step2}")

    # 서울에서 최종 지점까지의 직선 거리
    final_distance = seoul.distance_to(step2)
    print(f"서울에서 최종 지점까지 직선 거리: {float(final_distance):.2f}m")
    print(f"이론적 거리 (피타고라스): {(1000**2 + 1000**2) ** 0.5:.2f}m")

except Exception as e:
    print(f"연속 이동 테스트 오류: {e}")
    import traceback

    traceback.print_exc()

print("\n=== forward 메서드 테스트 완료 ===")
```

```python
# Degree 나눗셈 문제 분석
from math import pi

import dronesim.unit as unit

print("=== Degree 나눗셈 문제 분석 ===")

degree_90 = unit.Degree(90.0)
degree_45 = unit.Degree(45.0)

print(f"degree_90: {degree_90}")
print(f"degree_45: {degree_45}")

# 내부값 (라디안) 확인
print("\n=== 내부값 (라디안) ===")
print(f"degree_90 내부값: {float(degree_90):.10f} 라디안")
print(f"degree_45 내부값: {float(degree_45):.10f} 라디안")
print(f"90도 라디안: {90 * pi / 180:.10f}")
print(f"45도 라디안: {45 * pi / 180:.10f}")

# 나눗셈 계산
print("\n=== 나눗셈 분석 ===")
result = degree_90 / degree_45
print(f"90° / 45° = {result}")
print(f"결과 내부값: {float(result):.10f}")

# 예상 결과
expected_internal = float(degree_90) / float(degree_45)
print(f"예상 내부값 (라디안/라디안): {expected_internal:.10f}")

# 단위 변환으로 표시값 계산
expected_display = expected_internal * 180 / pi
print(f"예상 표시값 (도): {expected_display:.10f}")

print("\n=== 문제 발견 ===")
print(f"실제 결과: {float(result):.10f}")
print(f"예상 결과: {expected_internal:.10f} (= 2.0)")
print(f"차이: {abs(float(result) - expected_internal):.10f}")

# 다른 계산들도 확인
print("\n=== 다른 연산 확인 ===")
add_result = degree_90 + degree_45
sub_result = degree_90 - degree_45
mul_result = degree_90 * degree_45

print(f"90° + 45° = {add_result} (내부: {float(add_result):.6f})")
print(f"90° - 45° = {sub_result} (내부: {float(sub_result):.6f})")
print(f"90° * 45° = {mul_result} (내부: {float(mul_result):.6f})")

# 예상값과 비교
print("\n=== 예상값과 비교 ===")
print(f"덧셈 예상: {(90 + 45) * pi / 180:.6f} 라디안 = 135°")
print(f"뺄셈 예상: {(90 - 45) * pi / 180:.6f} 라디안 = 45°")
print("곱셈 예상: ??? (단위가 °²이 되어야 함)")
print(f"나눗셈 예상: {90 / 45:.6f} = 2.0 (무차원)")
```

```python
import time

from tqdm.notebook import tqdm

BAR_FMT = (
    "{l_bar}"                     # 좌측(설명 + 퍼센트)
    "{bar}"                       # 프로그레스 바
    "{r_bar}"                     # 우측(남은 시간, 속도, 개수)
)
# r_bar 기본 구성:  | {n_fmt}/{total_fmt} [{elapsed}<{remaining}, {rate_fmt}{postfix}]

pbar = tqdm(total=10_000, desc="Epoch 1", bar_format=BAR_FMT,
            colour="green",  # Jupyter/터미널 둘 다 색상 지원(최신 tqdm)
            ncols=0)         # 0이면 자동 너비

for i in range(10_000):
    # ... 작업 ...
    if i % 100 == 0:
        # postfix로 메트릭(손실/정확도/스루풋 등) 붙이기
        pbar.set_postfix(loss=f"{0.1234:.4f}", speed=f"{(i+1)/(pbar.format_dict['elapsed']+1e-9):.1f}/s")
    pbar.update(1); time.sleep(0.0005)

pbar.close()
```

```python
import re

import numpy as np
import pandas as pd

# ====== 설정 ======
CSV_PATH = "example/train.csv"     # <-- CSV 경로/파일명으로 바꿔주세요
TARGET_COL = "Time_taken(min)" # 스키마 그대로

# ====== 유틸 ======
def parse_minutes_col(s: pd.Series) -> pd.Series:
    """'30 (min)' 같은 문자열에서도 숫자만 추출해 분(float)으로."""
    if np.issubdtype(s.dtype, np.number):
        return pd.to_numeric(s, errors="coerce").astype(float)
    s = s.astype(str).str.strip()
    num = s.str.extract(r"([0-9]+(?:\.[0-9]+)?)", expand=False)
    return pd.to_numeric(num, errors="coerce").astype(float)

def parse_date_time(date_s: pd.Series, time_s: pd.Series) -> pd.Series:
    """Order_Date와 시각 문자열(Time_Orderd 등)을 합쳐 UTC-naive로 파싱."""
    ds = date_s.astype(str).str.strip()
    ts = time_s.astype(str).str.strip()
    # 시간 결측/이상치 보호
    ts = ts.where(ts.str.len() > 0, np.nan)
    dt = pd.to_datetime(ds + " " + ts, errors="coerce", utc=True)
    return dt

def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dphi, dlmb = lat2 - lat1, lon2 - lon1
    a = np.sin(dphi/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlmb/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

def safe_corr(a, b):
    m = np.isfinite(a) & np.isfinite(b)
    if m.sum() < 3:
        return np.nan
    return float(np.corrcoef(a[m], b[m])[0,1])

# ====== 로드 & 기본 전처리 ======
df = pd.read_csv(CSV_PATH)

# 필수 컬럼 확인
needed = {
    "Order_Date", "Time_Orderd", "Time_Order_picked",
    "Restaurant_latitude", "Restaurant_longitude",
    "Delivery_location_latitude", "Delivery_location_longitude",
    TARGET_COL,
}
missing = [c for c in needed if c not in df.columns]
if missing:
    raise SystemExit(f"누락 컬럼: {missing}")

# 타깃 파싱(분)
y = parse_minutes_col(df[TARGET_COL])

# 준비시간(분)
order_at  = parse_date_time(df["Order_Date"], df["Time_Orderd"])
picked_at = parse_date_time(df["Order_Date"], df["Time_Order_picked"])
prep_min  = (picked_at - order_at).dt.total_seconds() / 60.0

# 거리(지오데식, km)
geo_km = haversine_km(
    pd.to_numeric(df["Restaurant_latitude"], errors="coerce"),
    pd.to_numeric(df["Restaurant_longitude"], errors="coerce"),
    pd.to_numeric(df["Delivery_location_latitude"], errors="coerce"),
    pd.to_numeric(df["Delivery_location_longitude"], errors="coerce"),
)

# ====== 두 가설 비교 ======
# H1: TARGET = 주문→드롭오프(total)  ⇒ service_est = TARGET - prep
service_est_min = y - prep_min
# H2: TARGET = 픽업→드롭오프(service) ⇒ service_true = TARGET
service_true_min = y.copy()

# 각 가설의 물리적/통계적 지표
def hypothesis_report(name, service_min, note=""):
    m = np.isfinite(service_min) & (service_min > 0) & np.isfinite(geo_km)
    if m.sum() == 0:
        return {
            "name": name, "valid_rows": 0, "median_v": np.nan, "p95_v": np.nan,
            "neg_share": float(np.mean(service_min <= 0)) if np.isfinite(service_min).any() else np.nan,
            "corr_dist": np.nan, "note": note
        }
    v_kmh = geo_km[m] / (service_min[m] / 60.0)
    return {
        "name": name,
        "valid_rows": int(m.sum()),
        "median_v": float(np.nanmedian(v_kmh)),
        "p95_v": float(np.nanpercentile(v_kmh, 95)),
        "neg_share": float(np.mean(service_min <= 0)),                 # 전체에서 음수 비율
        "corr_dist": safe_corr(service_min, geo_km),                   # 시간-거리 상관
        "note": note
    }

rep_total  = hypothesis_report("H1: order→dropoff (TARGET)", service_est_min, note="TARGET - prep")
rep_service= hypothesis_report("H2: pickup→dropoff (TARGET)", service_true_min, note="TARGET 자체")

# 추가 지표: prep 대비 관계
m2 = np.isfinite(y) & np.isfinite(prep_min)
viol_share = float(np.mean(y[m2] < prep_min[m2] - 1.0)) if m2.any() else np.nan  # 1분 이상 작으면 위반으로

# 결론 로직(간단 규칙식)
def judge(rep_total, rep_service, viol_share):
    # 1) H1에서 음수 비율 높으면(>10%), total 가설 탈락
    # 2) H2의 속도 분포가 현실적(대략 median 15~45, p95 ≤ 80)이고 거리 상관이 더 높으면 H2 채택
    score_total = 0
    score_service = 0

    if not np.isnan(viol_share) and viol_share > 0.1:
        score_service += 1   # prep보다 작은 TARGET이 많다 → 서비스시간일 가능성↑
    if rep_total["neg_share"] > 0.1:
        score_service += 1   # total 가설에서 음수가 많이 나옴 → 말이 안 됨
    # 속도 현실성
    def plausible(rep):
        return (not np.isnan(rep["median_v"])) and (15 <= rep["median_v"] <= 45) and (rep["p95_v"] <= 80)
    if plausible(rep_service) and not plausible(rep_total):
        score_service += 1
    if rep_service["corr_dist"] > rep_total["corr_dist"]:
        score_service += 1

    concl = "pickup→dropoff (서비스 시간)" if score_service >= 2 else "order→dropoff (총 소요)"
    conf = "HIGH" if score_service >= 3 else ("MEDIUM" if score_service == 2 else "LOW")
    return concl, conf

conclusion, confidence = judge(rep_total, rep_service, viol_share)

# ====== 출력 ======
print("\n[Summary]")
print(f"rows                = {len(df)}")
print(f"usable target rows  = {np.isfinite(y).sum()}")
print(f"prep available rows = {np.isfinite(prep_min).sum()}")
print(f"geo_km available    = {np.isfinite(geo_km).sum()}")

print("\n[Hypothesis metrics]")
for rep in (rep_total, rep_service):
    print(f"- {rep['name']:<34s} | valid={rep['valid_rows']:>6d} | "
          f"median_v={rep['median_v']:.1f} km/h | p95_v={rep['p95_v']:.1f} km/h | "
          f"neg_share={rep['neg_share']:.2%} | corr(time,dist)={rep['corr_dist']:.3f} | {rep['note']}")

print(f"\n[Prep vs Target] share(TARGET < prep - 1min) = {viol_share:.2%}")

print(f"\n[Conclusion] TARGET is likely **{conclusion}**  (confidence: {confidence})")
```

```python
# --- (권장) parse_date_time에 dayfirst=True 패치 ---
# parse_date_time(...) 정의에 utc=True, dayfirst=True를 사용하세요.
# (이미 위에서 함수가 정의돼 있다면, 아래 한 줄로 재정의해도 됩니다)
def parse_date_time(date_s: pd.Series, time_s: pd.Series) -> pd.Series:
    ds = date_s.astype(str).str.strip()
    ts = time_s.astype(str).str.strip().replace({"": np.nan, "NaN": np.nan, "nan": np.nan})
    return pd.to_datetime(ds + " " + ts, errors="coerce", utc=True, dayfirst=True)

# --- 시간대(3시간 bin) 만들기: 기본은 '주문시각', 없으면 '픽업시각' ---
order_at  = parse_date_time(df["Order_Date"], df["Time_Orderd"])
picked_at = parse_date_time(df["Order_Date"], df["Time_Order_picked"])
time_base = picked_at.where(picked_at.notna(), order_at)  # pickup 우선, 없으면 order

# 시각 결측 보호
bin_valid = time_base.notna()
bin_hour = time_base.dt.hour.where(bin_valid, np.nan)
bin_start = (bin_hour // 3) * 3  # 0,3,6,...,21
bin_label = bin_start.map(lambda h: f"{int(h):02d}-{int((h+3)%24):02d}" if np.isfinite(h) else "unknown")
df["_bin3h"] = bin_label

# --- '비현실/불가능' 정의 ---
# H2(서비스 가설): v = geo_km / (TARGET_min/60)
service_min = service_true_min  # 위에서 만든 y.copy()
v_service = geo_km / (service_min / 60.0)
sus_service = (
    (~np.isfinite(service_min)) | (service_min <= 0) |
    (~np.isfinite(v_service))   | (v_service > 80) | (v_service < 3)
)

# H1(총소요 가설): service_est = TARGET - prep 에 동일 기준 적용
service_est_min = service_est_min  # 위에서 계산됨
v_total_as_service = geo_km / (service_est_min / 60.0)
sus_total = (
    (~np.isfinite(service_est_min)) | (service_est_min <= 0) |
    (~np.isfinite(v_total_as_service)) | (v_total_as_service > 80) | (v_total_as_service < 3)
)

# --- 3시간 bin 집계 ---
order_bins = [f"{h:02d}-{(h+3)%24:02d}" for h in range(0, 24, 3)]
summary_3h = (
    pd.DataFrame({
        "bin": df["_bin3h"],
        "sus_service": sus_service,
        "sus_total": sus_total,
    })
    .groupby("bin", dropna=False)
    .agg(total=("sus_service", "size"),
         sus_service=("sus_service", "sum"),
         sus_total=("sus_total", "sum"))
    .assign(
        rate_service=lambda d: d["sus_service"] / d["total"],
        rate_total=lambda d: d["sus_total"] / d["total"],
    )
    .reindex(order_bins + (["unknown"] if "unknown" in df["_bin3h"].unique() else []))
)

print("\n[Impossible/Implausible by 3h-of-day]")
print(summary_3h.fillna(0).to_string())

# --- (선택) 트래픽 등과 교차표 ---
if "Road_traffic_density" in df.columns:
    cross = (
        pd.DataFrame({"bin": df["_bin3h"], "sus_service": sus_service, "traffic": df["Road_traffic_density"]})
        .pivot_table(index="bin", columns="traffic", values="sus_service", aggfunc="sum", fill_value=0)
        .reindex(order_bins + (["unknown"] if "unknown" in df["_bin3h"].unique() else []))
    )
    print("\n[Suspects (service) by 3h × traffic]")
    print(cross.to_string())

# --- (선택) 최상위 이상치 샘플 확인: 속도 높은 순 ---
top_bad = (
    pd.DataFrame({
        "idx": df.index,
        "bin": df["_bin3h"],
        "geo_km": geo_km,
        "service_min": service_min,
        "v_kmh": v_service,
        "sus_service": sus_service,
        "City": df.get("City"),
        "Road_traffic_density": df.get("Road_traffic_density"),
        "Festival": df.get("Festival"),
    })
    .loc[sus_service]
    .sort_values("v_kmh", ascending=False)
    .head(15)
)
print("\n[Top 15 suspect rows (service) by highest v_kmh]")
print(top_bad.to_string(index=False))
```

```python
# === ROI(반경 50 km)에서 식당→배송지 거리 분포 ===
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ---- df가 이미 있으면 사용, 없으면 CSV 로드 ----
try:
    assert isinstance(df, pd.DataFrame)
except NameError:
    CSV_PATH = "example/train.csv"   # 필요시 경로 수정
    df = pd.read_csv(CSV_PATH)

# ---- 컬럼명 지정 ----
LAT_R, LON_R = "Restaurant_latitude", "Restaurant_longitude"
LAT_D, LON_D = "Delivery_location_latitude", "Delivery_location_longitude"

# ---- 상수: 중심 경도/위도, 반경, 필터 모드 ----
CENTER_LAT = 12.959861         # ✅ 위도를 지정하지 않으면 데이터 평균 위도를 사용
CENTER_LON = 77.579225   # ✅ 너가 원하는 중심 경도를 여기에 적어라 (deg)
RADIUS_KM  = 50.0         # 반경(km)
MODE = "both"             # {"both", "either", "midpoint"}

# ---- 하버사인 거리(km) 유틸 ----
def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dφ = lat2 - lat1
    dλ = lon2 - lon1
    a = np.sin(dφ/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dλ/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

# ---- 좌표 로드/정리 ----
lat_r = pd.to_numeric(df[LAT_R], errors="coerce").to_numpy()
lon_r = pd.to_numeric(df[LON_R], errors="coerce").to_numpy()
lat_d = pd.to_numeric(df[LAT_D], errors="coerce").to_numpy()
lon_d = pd.to_numeric(df[LON_D], errors="coerce").to_numpy()

# 중심 위도 자동 결정(경도만 지정한 경우)
if CENTER_LAT is None:
    CENTER_LAT = float(np.nanmean(np.concatenate([lat_r, lat_d])))

# ---- 반경(50 km) ROI 마스크 만들기 ----
d_center_rest = haversine_km(lat_r, lon_r, CENTER_LAT, CENTER_LON)
d_center_dest = haversine_km(lat_d, lon_d, CENTER_LAT, CENTER_LON)

m_rest = np.isfinite(d_center_rest) & (d_center_rest <= RADIUS_KM)
m_dest = np.isfinite(d_center_dest) & (d_center_dest <= RADIUS_KM)

if MODE == "both":
    roi_mask = m_rest & m_dest
elif MODE == "either":
    roi_mask = m_rest | m_dest
elif MODE == "midpoint":
    lat_mid = (lat_r + lat_d) / 2.0
    lon_mid = (lon_r + lon_d) / 2.0
    d_mid   = haversine_km(lat_mid, lon_mid, CENTER_LAT, CENTER_LON)
    roi_mask = np.isfinite(d_mid) & (d_mid <= RADIUS_KM)
else:
    raise ValueError("MODE must be one of {'both','either','midpoint'}")

# ---- 식당→배송지 지오데식 거리(km) 계산 ----
dist_km = haversine_km(lat_r, lon_r, lat_d, lon_d)
valid   = roi_mask & np.isfinite(dist_km)
x = dist_km[valid]

print(f"[INFO] center=(lat={CENTER_LAT:.6f}, lon={CENTER_LON:.6f}), radius={RADIUS_KM} km, mode={MODE}")
print(f"[INFO] total rows={len(df):,}, ROI valid rows={valid.sum():,} ({valid.sum()/len(df)*100:.2f}%)")

# ---- 요약 통계 ----
if x.size == 0:
    print("No data in ROI. Adjust CENTER_/RADIUS/MODE.")
else:
    q = np.percentile(x, [5, 25, 50, 75, 90, 95])
    summary = {
        "count": int(x.size),
        "mean_km": float(np.mean(x)),
        "std_km": float(np.std(x, ddof=1)) if x.size > 1 else 0.0,
        "min_km": float(np.min(x)),
        "p05_km": float(q[0]),
        "p25_km": float(q[1]),
        "median_km": float(q[2]),
        "p75_km": float(q[3]),
        "p90_km": float(q[4]),
        "p95_km": float(q[5]),
        "max_km": float(np.max(x)),
    }
    print("[Summary within ROI (km)]")
    for k, v in summary.items():
        print(f"  {k:>10s}: {v:.3f}" if isinstance(v, float) else f"  {k:>10s}: {v}")

    # ---- 5 km bin 분포 표 ----
    BIN_W = 5.0
    upper = BIN_W * np.ceil(max(1.0, x.max()) / BIN_W)
    edges = np.arange(0.0, upper + BIN_W + 1e-9, BIN_W)
    labels = [f"{int(edges[i])}-{int(edges[i+1])} km" for i in range(len(edges)-1)]
    cats = pd.cut(x, bins=edges, right=False, labels=labels, include_lowest=True)
    tab = pd.value_counts(cats, sort=False).reindex(labels, fill_value=0)
    share = (tab / tab.sum() * 100).round(2)

    out = pd.DataFrame({"count": tab.to_numpy(), "share_pct": share.to_numpy()}, index=labels)
    print("\n[5 km bins within ROI]")
    print(out.to_string())

    # ---- 히스토그램 / CDF 그림 ----
    plt.figure()
    plt.hist(x, bins=edges)
    plt.title("Restaurant→Delivery distance (ROI)")
    plt.xlabel("Distance (km)")
    plt.ylabel("Count")
    plt.grid(True, linewidth=0.5, alpha=0.5)
    plt.tight_layout()

    plt.figure()
    xs = np.sort(x)
    ys = np.arange(1, xs.size + 1) / xs.size
    plt.plot(xs, ys, linewidth=2)
    plt.title("CDF of distance (ROI)")
    plt.xlabel("Distance (km)")
    plt.ylabel("Cumulative fraction")
    plt.grid(True, linewidth=0.5, alpha=0.5)
    plt.tight_layout()

    plt.show()
```

```python
# === ROI(반경 50 km)에서 식당→배송지 거리 분포: 교통 혼잡도별 시각화 (겹치지 않는 막대) ===
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# -------------------------
# 0) 데이터 로드/준비
# -------------------------
try:
    assert isinstance(df, pd.DataFrame)
except NameError:
    CSV_PATH = "example/train.csv"   # 필요 시 경로 수정
    df = pd.read_csv(CSV_PATH)

# 스키마 열 이름 (질문에 제공한 정확한 이름 사용)
COLS = {
    "lat_r": "Restaurant_latitude",
    "lon_r": "Restaurant_longitude",
    "lat_d": "Delivery_location_latitude",
    "lon_d": "Delivery_location_longitude",
    "traffic": "Road_traffic_density",
}
for k, v in COLS.items():
    if v not in df.columns:
        raise KeyError(f"Missing column: {v}")

# -------------------------
# 1) 상수 설정 (ROI)
# -------------------------
CENTER_LAT = 12.959861    # 필요 시 None → 자동 평균 위도
CENTER_LON = 77.579225    # 중심 경도
RADIUS_KM  = 50.0         # 반경(km)
MODE = "both"             # {"both", "either", "midpoint"}
BIN_W = 5.0               # 히스토그램 bin 폭(km)
MIN_DISTANCE = 0.0        # 음수/0 거리 제거용 하한(일반적으로 0.0)

# -------------------------
# 2) 유틸: 하버사인 거리(km)
# -------------------------
def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dφ = lat2 - lat1
    dλ = lon2 - lon1
    a = np.sin(dφ/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dλ/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

# -------------------------
# 3) 좌표/열 정리
# -------------------------
lat_r = pd.to_numeric(df[COLS["lat_r"]], errors="coerce").to_numpy()
lon_r = pd.to_numeric(df[COLS["lon_r"]], errors="coerce").to_numpy()
lat_d = pd.to_numeric(df[COLS["lat_d"]], errors="coerce").to_numpy()
lon_d = pd.to_numeric(df[COLS["lon_d"]], errors="coerce").to_numpy()

# 중심 위도 자동(경도만 고정하고 싶을 때)
if CENTER_LAT is None:
    CENTER_LAT = float(np.nanmean(np.concatenate([lat_r, lat_d])))

# -------------------------
# 4) ROI 마스크 계산
# -------------------------
d_center_rest = haversine_km(lat_r, lon_r, CENTER_LAT, CENTER_LON)
d_center_dest = haversine_km(lat_d, lon_d, CENTER_LAT, CENTER_LON)

m_rest = np.isfinite(d_center_rest) & (d_center_rest <= RADIUS_KM)
m_dest = np.isfinite(d_center_dest) & (d_center_dest <= RADIUS_KM)

if MODE == "both":
    roi_mask = m_rest & m_dest
elif MODE == "either":
    roi_mask = m_rest | m_dest
elif MODE == "midpoint":
    lat_mid = (lat_r + lat_d) / 2.0
    lon_mid = (lon_r + lon_d) / 2.0
    d_mid   = haversine_km(lat_mid, lon_mid, CENTER_LAT, CENTER_LON)
    roi_mask = np.isfinite(d_mid) & (d_mid <= RADIUS_KM)
else:
    raise ValueError("MODE must be one of {'both','either','midpoint'}")

# -------------------------
# 5) 식당→배송지 거리(km)
# -------------------------
dist_km = haversine_km(lat_r, lon_r, lat_d, lon_d)
valid_dist = np.isfinite(dist_km) & (dist_km > MIN_DISTANCE)

# -------------------------
# 6) 교통 혼잡도 정규화 (문자열 정리)
# -------------------------
traffic_raw = df[COLS["traffic"]].astype("string")

def norm_traffic(x: pd.Series) -> pd.Series:
    x = x.str.strip().str.lower()
    # 흔한 표기 정규화
    mapping = {
        "low": "low", "low ": "low",
        "medium": "medium", "moderate": "medium",
        "high": "high",
        "jam": "jam", "traffic jam": "jam",
        "nan": np.nan, "": np.nan, "none": np.nan, "unknown": np.nan
    }
    y = x.map(mapping).fillna(x)  # 매핑 외 값은 원문 유지
    # 최종 카테고리 집합 정리
    return y.where(y.isin(["low","medium","high","jam"]), other="other")

traffic = norm_traffic(traffic_raw)

# -------------------------
# 7) 최종 유효 마스크 & 데이터프레임 구성
# -------------------------
valid = roi_mask & valid_dist & traffic.notna().to_numpy()
data = pd.DataFrame({
    "distance_km": dist_km[valid],
    "traffic": traffic[valid].to_numpy()
})

print(f"[INFO] center=(lat={CENTER_LAT:.6f}, lon={CENTER_LON:.6f}), "
      f"radius={RADIUS_KM} km, mode={MODE}")
print(f"[INFO] total rows={len(df):,}, ROI+valid rows={len(data):,} "
      f"({(len(data)/len(df)*100 if len(df)>0 else 0):.2f}%)")

if data.empty:
    print("No data in ROI. Adjust CENTER_/RADIUS/MODE or check columns.")
else:
    # -------------------------
    # 8) 공통 bin 경계 (모든 그래프에서 공유)
    # -------------------------
    upper = BIN_W * np.ceil(max(1.0, data["distance_km"].max()) / BIN_W)
    edges = np.arange(0.0, upper + BIN_W + 1e-9, BIN_W)
    centers = edges[:-1] + BIN_W / 2

    # -------------------------
    # 8-1) 교통 혼잡도별 히스토그램 (겹치지 않게 side-by-side)
    # -------------------------
    cats_all = ["low","medium","high","jam","other"]
    present = [c for c in cats_all if (data["traffic"]==c).any()]

    # 색상 팔레트
    colors = plt.cm.tab10.colors[:len(present)]

    # 각 카테고리별 빈도 계산
    counts = []
    for c in present:
        vals = data.loc[data["traffic"]==c, "distance_km"].to_numpy()
        hist, _ = np.histogram(vals, bins=edges)
        counts.append(hist)
    counts = np.array(counts)   # shape = (n_categories, n_bins)

    width = BIN_W / (len(present) + 1)  # 막대 너비 축소 (겹침 방지)

    plt.figure(figsize=(9,5))
    for i, c in enumerate(present):
        plt.bar(centers + (i - len(present)/2)*width + width/2,
                counts[i],
                width=width,
                label=f"{c} (n={counts[i].sum()})",
                color=colors[i])
    plt.title(f"Restaurant→Delivery Distance by Road Traffic (ROI={RADIUS_KM} km)")
    plt.xlabel("Distance (km)")
    plt.ylabel("Count")
    plt.grid(True, axis='y', linewidth=0.5, alpha=0.5)
    plt.legend()
    plt.tight_layout()

    # -------------------------
    # 8-2) 교통 혼잡도별 CDF
    # -------------------------
    plt.figure(figsize=(8,5))
    for i, c in enumerate(present):
        x = np.sort(data.loc[data["traffic"]==c, "distance_km"].to_numpy())
        if x.size == 0:
            continue
        y = np.arange(1, x.size+1) / x.size
        plt.plot(x, y, linewidth=2, label=f"{c} (n={x.size})")
    plt.title("CDF of Restaurant→Delivery Distance by Road Traffic")
    plt.xlabel("Distance (km)")
    plt.ylabel("Cumulative fraction")
    plt.grid(True, linewidth=0.5, alpha=0.5)
    plt.legend()
    plt.tight_layout()

    # -------------------------
    # 8-3) 교통 혼잡도별 박스플롯
    # -------------------------
    plt.figure(figsize=(7,5))
    grouped = [data.loc[data["traffic"]==c, "distance_km"].to_numpy() for c in present]
    plt.boxplot(grouped, labels=present, showfliers=False)
    plt.title("Distance Distribution by Road Traffic (Boxplot)")
    plt.ylabel("Distance (km)")
    plt.grid(True, axis='y', linewidth=0.5, alpha=0.5)
    plt.tight_layout()

    # -------------------------
    # 8-4) 5 km bin 분포 표 (혼잡도별)
    # -------------------------
    print("\n[5 km bins within ROI by Road Traffic]")
    bin_labels = [f"{int(edges[i])}-{int(edges[i+1])} km" for i in range(len(edges)-1)]
    for c in present:
        x = data.loc[data["traffic"]==c, "distance_km"].to_numpy()
        cats_cut = pd.cut(x, bins=edges, right=False, labels=bin_labels, include_lowest=True)
        tab = pd.value_counts(cats_cut, sort=False).reindex(bin_labels, fill_value=0)
        share = (tab / max(tab.sum(), 1) * 100).round(2)
        out = pd.DataFrame({"count": tab.to_numpy(), "share_pct": share.to_numpy()},
                           index=bin_labels)
        print(f"\n== {c.upper()} ==")
        print(out.to_string())

    plt.show()
```

```python
# === ROI(반경 50 km)에서 식당→배송지 '거리' 분포: City × Road_traffic_density 동시 분석 ===
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ------------------------------------------------------------
# 0) 데이터 로드/준비
# ------------------------------------------------------------
try:
    assert isinstance(df, pd.DataFrame)
except NameError:
    CSV_PATH = "example/train.csv"   # 필요 시 경로 수정
    df = pd.read_csv(CSV_PATH)

# 스키마 열 이름 (질문에서 준 이름 사용)
COLS = {
    "lat_r": "Restaurant_latitude",
    "lon_r": "Restaurant_longitude",
    "lat_d": "Delivery_location_latitude",
    "lon_d": "Delivery_location_longitude",
    "city":  "City",
    "traffic": "Road_traffic_density",
}
for k, v in COLS.items():
    if v not in df.columns:
        raise KeyError(f"Missing column: {v}")

# ------------------------------------------------------------
# 1) 파라미터 (원형 ROI & 플로팅 설정)
# ------------------------------------------------------------
CENTER_LAT = 12.959861     # 필요시 None → 자동 평균 위도
CENTER_LON = 77.579225     # 중심 경도
RADIUS_KM  = 50.0          # 반경(km)
MODE = "both"              # {"both","either","midpoint"}
BIN_W = 5.0                # 히스토그램 bin 폭(km)
MIN_DISTANCE = 0.0         # 음수/0 제거 하한
TRAFFIC_MODE = "four"      # {"four","binary"} (four=low/medium/high/jam, binary=free/congested)

# ------------------------------------------------------------
# 2) 유틸: 하버사인 거리(km)
# ------------------------------------------------------------
def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dφ = lat2 - lat1
    dλ = lon2 - lon1
    a = np.sin(dφ/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dλ/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

# ------------------------------------------------------------
# 3) 좌표/열 정리
# ------------------------------------------------------------
lat_r = pd.to_numeric(df[COLS["lat_r"]], errors="coerce").to_numpy()
lon_r = pd.to_numeric(df[COLS["lon_r"]], errors="coerce").to_numpy()
lat_d = pd.to_numeric(df[COLS["lat_d"]], errors="coerce").to_numpy()
lon_d = pd.to_numeric(df[COLS["lon_d"]], errors="coerce").to_numpy()

if CENTER_LAT is None:
    CENTER_LAT = float(np.nanmean(np.concatenate([lat_r, lat_d])))

# ------------------------------------------------------------
# 4) ROI 마스크 계산
# ------------------------------------------------------------
d_center_rest = haversine_km(lat_r, lon_r, CENTER_LAT, CENTER_LON)
d_center_dest = haversine_km(lat_d, lon_d, CENTER_LAT, CENTER_LON)

m_rest = np.isfinite(d_center_rest) & (d_center_rest <= RADIUS_KM)
m_dest = np.isfinite(d_center_dest) & (d_center_dest <= RADIUS_KM)

if MODE == "both":
    roi_mask = m_rest & m_dest
elif MODE == "either":
    roi_mask = m_rest | m_dest
elif MODE == "midpoint":
    lat_mid = (lat_r + lat_d) / 2.0
    lon_mid = (lon_r + lon_d) / 2.0
    d_mid   = haversine_km(lat_mid, lon_mid, CENTER_LAT, CENTER_LON)
    roi_mask = np.isfinite(d_mid) & (d_mid <= RADIUS_KM)
else:
    raise ValueError("MODE must be one of {'both','either','midpoint'}")

# ------------------------------------------------------------
# 5) 식당→배송지 '거리'(km)
# ------------------------------------------------------------
dist_km = haversine_km(lat_r, lon_r, lat_d, lon_d)
valid_dist = np.isfinite(dist_km) & (dist_km > MIN_DISTANCE)

# ------------------------------------------------------------
# 6) City / Road_traffic_density 정규화
# ------------------------------------------------------------
city_raw = df[COLS["city"]].astype("string")
traffic_raw = df[COLS["traffic"]].astype("string")

def norm_city(x: pd.Series) -> pd.Series:
    x = x.str.strip().str.lower()
    mapping = {
        "metropolitan": "metropolitan", "metropolitian": "metropolitan",
        "metro": "metropolitan", "metro city": "metropolitan",
        "urban": "urban", "urb": "urban",
        "semi-urban": "semi-urban", "semi urban": "semi-urban", "suburban": "semi-urban",
        "nan": np.nan, "": np.nan, "none": np.nan, "unknown": np.nan,
    }
    y = x.map(mapping).fillna(x)
    return y.where(y.isin(["metropolitan","urban","semi-urban"]), other="other")

def norm_traffic(x: pd.Series) -> pd.Series:
    x = x.str.strip().str.lower()
    mapping = {
        "low": "low", "low ": "low",
        "medium": "medium", "moderate": "medium",
        "high": "high",
        "jam": "jam", "traffic jam": "jam",
        "nan": np.nan, "": np.nan, "none": np.nan, "unknown": np.nan
    }
    y = x.map(mapping).fillna(x)
    return y.where(y.isin(["low","medium","high","jam"]), other="other")

city = norm_city(city_raw)
traffic4 = norm_traffic(traffic_raw)

def binarize_traffic(s: pd.Series) -> pd.Series:
    # free: low/medium, congested: high/jam, 나머지는 NaN
    return pd.Series(
        np.where(s.isin(["low","medium"]), "free",
        np.where(s.isin(["high","jam"]), "congested", np.nan)),
        index=s.index
    ).astype("string")

if TRAFFIC_MODE == "binary":
    traffic_cat = binarize_traffic(traffic4)
    traffic_levels = ["free","congested"]
    pretty_tr = {"free":"Free", "congested":"Congested"}
elif TRAFFIC_MODE == "four":
    traffic_cat = traffic4
    traffic_levels = ["low","medium","high","jam"]
    pretty_tr = {"low":"Low","medium":"Medium","high":"High","jam":"Jam"}
else:
    raise ValueError("TRAFFIC_MODE must be 'four' or 'binary'")

# ------------------------------------------------------------
# 7) 최종 유효 필터 & 분석용 DF (City: Metropolitan vs Urban만)
# ------------------------------------------------------------
target_cities = ["metropolitan","urban"]

valid = (
    roi_mask &
    valid_dist &
    city.isin(target_cities).to_numpy() &
    traffic_cat.isin(traffic_levels).to_numpy()
)

data = pd.DataFrame({
    "distance_km": dist_km[valid],
    "city": city[valid].to_numpy(),
    "traffic": traffic_cat[valid].to_numpy(),
})

print(f"[INFO] center=(lat={CENTER_LAT:.6f}, lon={CENTER_LON:.6f}), radius={RADIUS_KM} km, mode={MODE}")
print(f"[INFO] total rows={len(df):,}, ROI+valid rows={len(data):,} "
      f"({(len(data)/len(df)*100 if len(df)>0 else 0):.2f}%)")

if data.empty:
    print("No data in ROI (or no Metropolitan/Urban with chosen traffic levels).")
else:
    # ------------------------------------------------------------
    # 8) 공통 bin 경계
    # ------------------------------------------------------------
    upper = BIN_W * np.ceil(max(1.0, data["distance_km"].max()) / BIN_W)
    edges = np.arange(0.0, upper + BIN_W + 1e-9, BIN_W)
    centers = edges[:-1] + BIN_W/2
    bin_labels = [f"{int(edges[i])}-{int(edges[i+1])} km" for i in range(len(edges)-1)]

    city_label = {"metropolitan":"Metropolitan", "urban":"Urban"}

    # ------------------------------------------------------------
    # 8-1) 히스토그램 (City별 Figure, 내부는 traffic 여러 막대 / side-by-side)
    # ------------------------------------------------------------
    for cty in [c for c in target_cities if (data["city"]==c).any()]:
        sub = data[data["city"]==cty]
        present_tr = [t for t in traffic_levels if (sub["traffic"]==t).any()]
        if not present_tr:
            continue

        counts = []
        for t in present_tr:
            vals = sub.loc[sub["traffic"]==t, "distance_km"].to_numpy()
            hist, _ = np.histogram(vals, bins=edges)
            counts.append(hist)
        counts = np.array(counts)  # (n_tr, n_bins)

        width = BIN_W / (len(present_tr) + 1)
        colors = plt.cm.Set2.colors[:len(present_tr)]

        plt.figure(figsize=(9,5))
        for i, t in enumerate(present_tr):
            plt.bar(centers + (i - len(present_tr)/2)*width + width/2,
                    counts[i],
                    width=width,
                    label=f"{pretty_tr[t]} (n={counts[i].sum()})",
                    color=colors[i])
        plt.title(f"[{city_label[cty]}] Restaurant→Delivery Distance by Traffic (ROI={RADIUS_KM} km)")
        plt.xlabel("Distance (km)")
        plt.ylabel("Count")
        plt.grid(True, axis='y', linewidth=0.5, alpha=0.5)
        plt.legend()
        plt.tight_layout()

    # ------------------------------------------------------------
    # 8-2) CDF (City별 Figure, traffic별 선)
    # ------------------------------------------------------------
    for cty in [c for c in target_cities if (data["city"]==c).any()]:
        sub = data[data["city"]==cty]
        present_tr = [t for t in traffic_levels if (sub["traffic"]==t).any()]
        if not present_tr:
            continue

        plt.figure(figsize=(8,5))
        for t in present_tr:
            x = np.sort(sub.loc[sub["traffic"]==t, "distance_km"].to_numpy())
            if x.size == 0:
                continue
            y = np.arange(1, x.size+1)/x.size
            plt.plot(x, y, linewidth=2, label=f"{pretty_tr[t]} (n={x.size})")
        plt.title(f"[{city_label[cty]}] CDF of Distance by Traffic")
        plt.xlabel("Distance (km)")
        plt.ylabel("Cumulative fraction")
        plt.grid(True, linewidth=0.5, alpha=0.5)
        plt.legend()
        plt.tight_layout()

    # ------------------------------------------------------------
    # 8-3) 박스플롯 (City별 Figure, traffic별 박스)
    # ------------------------------------------------------------
    for cty in [c for c in target_cities if (data["city"]==c).any()]:
        sub = data[data["city"]==cty]
        present_tr = [t for t in traffic_levels if (sub["traffic"]==t).any()]
        if not present_tr:
            continue

        grouped = [sub.loc[sub["traffic"]==t, "distance_km"].to_numpy() for t in present_tr]
        labels  = [pretty_tr[t] for t in present_tr]

        plt.figure(figsize=(7,5))
        plt.boxplot(grouped, labels=labels, showfliers=False)
        plt.title(f"[{city_label[cty]}] Distance Distribution by Traffic (Boxplot)")
        plt.ylabel("Distance (km)")
        plt.grid(True, axis='y', linewidth=0.5, alpha=0.5)
        plt.tight_layout()

    # ------------------------------------------------------------
    # 8-4) 5 km bin 분포 표 (City × Traffic)
    # ------------------------------------------------------------
    print("\n[5 km bins within ROI by City × Traffic]")
    for cty in [c for c in target_cities if (data["city"]==c).any()]:
        sub = data[data["city"]==cty]
        present_tr = [t for t in traffic_levels if (sub["traffic"]==t).any()]
        if not present_tr:
            continue

        print(f"\n== {city_label[cty].upper()} ==")
        for t in present_tr:
            x = sub.loc[sub["traffic"]==t, "distance_km"].to_numpy()
            cats_cut = pd.cut(x, bins=edges, right=False, labels=bin_labels, include_lowest=True)
            tab = pd.value_counts(cats_cut, sort=False).reindex(bin_labels, fill_value=0)
            share = (tab / max(tab.sum(),1) * 100).round(2)
            out = pd.DataFrame({"count": tab.to_numpy(), "share_pct": share.to_numpy()},
                               index=bin_labels)
            print(f"\n-- {pretty_tr[t]} --")
            print(out.to_string())

    # ------------------------------------------------------------
    # 8-5) 요약 통계 (City × Traffic)
    # ------------------------------------------------------------
    summary = (
        data
        .groupby(["city","traffic"])["distance_km"]
        .agg(count="count", mean_km="mean", median_km="median",
             p25_km=lambda s: np.percentile(s,25),
             p75_km=lambda s: np.percentile(s,75),
             std_km=lambda s: s.std(ddof=1) if s.size>1 else 0.0,
             min_km="min", max_km="max")
        .reset_index()
    )
    # 보기 좋게 라벨 바꾸기
    summary["city"] = summary["city"].map(city_label)
    summary["traffic"] = summary["traffic"].map(pretty_tr)
    print("\n[Summary stats of distance (km) by City × Traffic]")
    print(summary.to_string(index=False))

    # ------------------------------------------------------------
    # (옵션) 8-6) 2-way ANOVA: distance ~ City + Traffic + City:Traffic
    # ------------------------------------------------------------
    try:
        import statsmodels.formula.api as smf
        model = smf.ols("distance_km ~ C(city) * C(traffic)", data=data).fit(cov_type="HC3")
        print("\n[Two-way OLS with interaction]")
        print(model.summary())
        try:
            from statsmodels.stats.anova import anova_lm
            print("\n[ANOVA table]")
            print(anova_lm(model, typ=2))
        except Exception as _:
            pass
    except Exception as _:
        print("\n[INFO] statsmodels가 없어서 OLS/ANOVA는 건너뜀. `pip install statsmodels` 후 재실행하세요.")

    plt.show()
```

```python
# === ROI(반경 50 km)에서 식당→배송지 '거리'를 Time_Orderd(주문시각)과 함께 분석: City × Traffic × Time ===
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ------------------------------------------------------------
# 0) 데이터 로드/준비
# ------------------------------------------------------------
try:
    assert isinstance(df, pd.DataFrame)
except NameError:
    CSV_PATH = "example/train.csv"   # 필요 시 경로 수정
    df = pd.read_csv(CSV_PATH)

# 스키마 열 이름 (질문에서 준 이름 사용)
COLS = {
    "lat_r": "Restaurant_latitude",
    "lon_r": "Restaurant_longitude",
    "lat_d": "Delivery_location_latitude",
    "lon_d": "Delivery_location_longitude",
    "city":  "City",
    "traffic": "Road_traffic_density",
    "t_ordered": "Time_Orderd",   # 주문 시각(문자열, 예: '16:30:12')
}
for k, v in COLS.items():
    if v not in df.columns:
        raise KeyError(f"Missing column: {v}")

# ------------------------------------------------------------
# 1) 파라미터 (원형 ROI & 플로팅/시간 bin 설정)
# ------------------------------------------------------------
CENTER_LAT = 12.959861     # 필요시 None → 자동 평균 위도
CENTER_LON = 77.579225     # 중심 경도
RADIUS_KM  = 50.0          # 반경(km)
MODE = "both"              # {"both","either","midpoint"}
BIN_W = 5.0                # 거리 히스토그램 bin 폭(km)
MIN_DISTANCE = 0.0         # 음수/0 제거 하한

TRAFFIC_MODE = "four"      # {"four","binary"} (four=low/medium/high/jam, binary=free/congested)

TIME_BIN_MIN = 60          # 시간 bin 폭(분). 60=시간별, 30=30분 구간 등
TIME_START_MIN = 0         # 0=00:00
TIME_END_MIN   = 24*60     # 1440=24:00

# ------------------------------------------------------------
# 2) 유틸: 하버사인 거리(km)
# ------------------------------------------------------------
def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dφ = lat2 - lat1
    dλ = lon2 - lon1
    a = np.sin(dφ/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dλ/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

# ------------------------------------------------------------
# 3) 좌표/열 정리 + Time_Orderd 파싱
# ------------------------------------------------------------
lat_r = pd.to_numeric(df[COLS["lat_r"]], errors="coerce").to_numpy()
lon_r = pd.to_numeric(df[COLS["lon_r"]], errors="coerce").to_numpy()
lat_d = pd.to_numeric(df[COLS["lat_d"]], errors="coerce").to_numpy()
lon_d = pd.to_numeric(df[COLS["lon_d"]], errors="coerce").to_numpy()

# 중심 위도 자동(경도만 고정하고 싶을 때)
if CENTER_LAT is None:
    CENTER_LAT = float(np.nanmean(np.concatenate([lat_r, lat_d])))

# 주문시각 파싱(시간만 있는 문자열을 '1900-01-01 ' 접두로 붙여 datetime으로 변환)
t_raw = df[COLS["t_ordered"]].astype("string")
t_parsed = pd.to_datetime(("1900-01-01 " + t_raw.fillna("")).replace("<NA>",""),
                          errors="coerce", format=None)

# 분 단위(0~1440), 시(hour) 도출
t_min = (t_parsed.dt.hour.fillna(0)*60 +
         t_parsed.dt.minute.fillna(0) +
         t_parsed.dt.second.fillna(0)/60.0)

t_hour = t_parsed.dt.hour.astype("Int64")  # 0..23 (결측은 <NA>)

# ------------------------------------------------------------
# 4) ROI 마스크 계산
# ------------------------------------------------------------
d_center_rest = haversine_km(lat_r, lon_r, CENTER_LAT, CENTER_LON)
d_center_dest = haversine_km(lat_d, lon_d, CENTER_LAT, CENTER_LON)

m_rest = np.isfinite(d_center_rest) & (d_center_rest <= RADIUS_KM)
m_dest = np.isfinite(d_center_dest) & (d_center_dest <= RADIUS_KM)

if MODE == "both":
    roi_mask = m_rest & m_dest
elif MODE == "either":
    roi_mask = m_rest | m_dest
elif MODE == "midpoint":
    lat_mid = (lat_r + lat_d) / 2.0
    lon_mid = (lon_r + lon_d) / 2.0
    d_mid   = haversine_km(lat_mid, lon_mid, CENTER_LAT, CENTER_LON)
    roi_mask = np.isfinite(d_mid) & (d_mid <= RADIUS_KM)
else:
    raise ValueError("MODE must be one of {'both','either','midpoint'}")

# ------------------------------------------------------------
# 5) 식당→배송지 '거리'(km)
# ------------------------------------------------------------
dist_km = haversine_km(lat_r, lon_r, lat_d, lon_d)
valid_dist = np.isfinite(dist_km) & (dist_km > MIN_DISTANCE)

# ------------------------------------------------------------
# 6) City / Road_traffic_density 정규화 + (옵션)이진화
# ------------------------------------------------------------
city_raw = df[COLS["city"]].astype("string")
traffic_raw = df[COLS["traffic"]].astype("string")

def norm_city(x: pd.Series) -> pd.Series:
    x = x.str.strip().str.lower()
    mapping = {
        "metropolitan": "metropolitan", "metropolitian": "metropolitan",
        "metro": "metropolitan", "metro city": "metropolitan",
        "urban": "urban", "urb": "urban",
        "semi-urban": "semi-urban", "semi urban": "semi-urban", "suburban": "semi-urban",
        "nan": np.nan, "": np.nan, "none": np.nan, "unknown": np.nan,
    }
    y = x.map(mapping).fillna(x)
    return y.where(y.isin(["metropolitan","urban","semi-urban"]), other="other")

def norm_traffic(x: pd.Series) -> pd.Series:
    x = x.str.strip().str.lower()
    mapping = {
        "low": "low", "low ": "low",
        "medium": "medium", "moderate": "medium",
        "high": "high",
        "jam": "jam", "traffic jam": "jam",
        "nan": np.nan, "": np.nan, "none": np.nan, "unknown": np.nan
    }
    y = x.map(mapping).fillna(x)
    return y.where(y.isin(["low","medium","high","jam"]), other="other")

city = norm_city(city_raw)
traffic4 = norm_traffic(traffic_raw)

def binarize_traffic(s: pd.Series) -> pd.Series:
    # free: low/medium, congested: high/jam, 나머지는 NaN
    return pd.Series(
        np.where(s.isin(["low","medium"]), "free",
        np.where(s.isin(["high","jam"]), "congested", np.nan)),
        index=s.index
    ).astype("string")

if TRAFFIC_MODE == "binary":
    traffic_cat = binarize_traffic(traffic4)
    traffic_levels = ["free","congested"]
    pretty_tr = {"free":"Free", "congested":"Congested"}
elif TRAFFIC_MODE == "four":
    traffic_cat = traffic4
    traffic_levels = ["low","medium","high","jam"]
    pretty_tr = {"low":"Low","medium":"Medium","high":"High","jam":"Jam"}
else:
    raise ValueError("TRAFFIC_MODE must be 'four' or 'binary'")

# ------------------------------------------------------------
# 7) 시간 구간(bin) 구성
# ------------------------------------------------------------
time_edges = np.arange(TIME_START_MIN, TIME_END_MIN + TIME_BIN_MIN, TIME_BIN_MIN, dtype=float)
time_centers = time_edges[:-1] + TIME_BIN_MIN/2
time_labels = [f"{int(e//60):02d}:{int(e%60):02d}-{int((e+TIME_BIN_MIN)//60)%24:02d}:{int((e+TIME_BIN_MIN)%60):02d}"
               for e in time_edges[:-1]]

# ------------------------------------------------------------
# 8) 최종 유효 필터 & 분석용 DF (City: Metropolitan/Urban)
# ------------------------------------------------------------
target_cities = ["metropolitan","urban"]

valid = (
    roi_mask &
    valid_dist &
    city.isin(target_cities).to_numpy() &
    traffic_cat.isin(traffic_levels).to_numpy() &
    pd.notna(t_parsed).to_numpy() &
    (t_min >= TIME_START_MIN).to_numpy() &
    (t_min < TIME_END_MIN).to_numpy()
)

data = pd.DataFrame({
    "distance_km": dist_km[valid],
    "city": city[valid].to_numpy(),
    "traffic": traffic_cat[valid].to_numpy(),
    "t_min": t_min[valid],
})
data["t_hour"] = (data["t_min"] // 60).astype(int)
data["t_bin"] = pd.cut(data["t_min"], bins=time_edges, right=False,
                       labels=time_labels, include_lowest=True)

print(f"[INFO] center=(lat={CENTER_LAT:.6f}, lon={CENTER_LON:.6f}), radius={RADIUS_KM} km, mode={MODE}")
print(f"[INFO] total rows={len(df):,}, ROI+valid rows={len(data):,} "
      f"({(len(data)/len(df)*100 if len(df)>0 else 0):.2f}%)")

if data.empty:
    print("No data in ROI (or parsing Time_Orderd failed). Check filters and time format.")
else:
    # ------------------------------------------------------------
    # 9) 시간대별 중앙값 거리 라인 그래프 (City별 figure, traffic별 선)
    # ------------------------------------------------------------
    city_label = {"metropolitan":"Metropolitan", "urban":"Urban"}
    for cty in [c for c in target_cities if (data["city"]==c).any()]:
        sub = data[data["city"]==cty]
        present_tr = [t for t in traffic_levels if (sub["traffic"]==t).any()]
        if not present_tr:
            continue

        # 시간대(hour) 프로파일: 중앙값(또는 평균) 거리
        prof = (
            sub.groupby(["traffic","t_hour"])["distance_km"]
               .median()  # 중앙값
               .unstack("traffic")
               .reindex(range(0,24))  # 0..23 정렬
        )

        plt.figure(figsize=(9,5))
        for t in present_tr:
            if t in prof.columns:
                plt.plot(prof.index, prof[t], linewidth=2, label=f"{pretty_tr[t]}")
        plt.title(f"[{city_label[cty]}] Median Distance by Order Hour (City × Traffic)")
        plt.xlabel("Order hour of day")
        plt.ylabel("Median distance (km)")
        plt.xticks(range(0,24,2))
        plt.grid(True, linewidth=0.5, alpha=0.5)
        plt.legend()
        plt.tight_layout()

    # ------------------------------------------------------------
    # 10) 시간×거리 2D 히트맵 (joint 분포) — City별 figure, traffic 합산
    #     (시간 bin × 거리 bin 카운트)
    # ------------------------------------------------------------
    # 거리 bin
    upper_d = BIN_W * np.ceil(max(1.0, data["distance_km"].max()) / BIN_W)
    dist_edges = np.arange(0.0, upper_d + BIN_W + 1e-9, BIN_W)
    dist_centers = dist_edges[:-1] + BIN_W/2

    for cty in [c for c in target_cities if (data["city"]==c).any()]:
        sub = data[data["city"]==cty]
        # 2D 히스토그램: rows=time_bins, cols=distance_bins
        # 시간: t_min in [0,1440)
        H, xedges, yedges = np.histogram2d(
            sub["t_min"].to_numpy(),
            sub["distance_km"].to_numpy(),
            bins=[time_edges, dist_edges]
        )
        plt.figure(figsize=(10,5))
        plt.imshow(H, origin="lower", aspect="auto",
                   extent=[dist_edges[0], dist_edges[-1], time_edges[0], time_edges[-1]])
        plt.colorbar(label="Count")
        plt.title(f"[{city_label[cty]}] Joint Distribution (Order time × Distance)")
        plt.xlabel("Distance (km)")
        plt.ylabel("Order time (minute of day)")
        yticks = np.arange(TIME_START_MIN, TIME_END_MIN+1, 120)  # 2시간 눈금
        plt.yticks(yticks, [f"{int(m//60):02d}:00" for m in yticks])
        plt.tight_layout()

    # ------------------------------------------------------------
    # 11) 요약 통계표 (City × Traffic × Time-bin)
    # ------------------------------------------------------------
    def p25(s): return np.percentile(s, 25)
    def p75(s): return np.percentile(s, 75)
    summary = (
        data.groupby(["city","traffic","t_bin"])["distance_km"]
            .agg(count="count", mean_km="mean", median_km="median",
                 p25_km=p25, p75_km=p75,
                 std_km=lambda s: s.std(ddof=1) if s.size>1 else 0.0)
            .reset_index()
    )
    # 보기 좋게 라벨 매핑
    summary["city"] = summary["city"].map(city_label)
    summary["traffic"] = summary["traffic"].map(lambda x: pretty_tr.get(x, x))
    print("\n[Summary: distance (km) by City × Traffic × Time-bin]")
    print(summary.to_string(index=False))

    # ------------------------------------------------------------
    # (옵션) 12) OLS: distance ~ City + Traffic + Hour + 상호작용 (강건표준오차)
    # ------------------------------------------------------------
    try:
        import statsmodels.formula.api as smf
        model = smf.ols(
            "distance_km ~ C(city) * C(traffic) + C(t_hour)",
            data=data.replace({"city": city_label, "traffic": pretty_tr})
        ).fit(cov_type="HC3")
        print("\n[OLS with interactions: distance ~ City * Traffic + Hour]")
        print(model.summary())
    except Exception as _:
        print("\n[INFO] statsmodels 미설치 또는 회귀 실패. `pip install statsmodels` 후 재실행하세요.")

    plt.show()
```

```python
# === Road Traffic Analysis by 30-min Intervals (ROI + 3AM adjustment) ===
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# -------------------------
# 0) Data Loading
# -------------------------
try:
    assert isinstance(df, pd.DataFrame)
except NameError:
    CSV_PATH = "example/train.csv"
    df = pd.read_csv(CSV_PATH)

COLS = {
    "traffic": "Road_traffic_density",
    "t_ordered": "Time_Orderd",
    "lat_r": "Restaurant_latitude",
    "lon_r": "Restaurant_longitude",
    "lat_d": "Delivery_location_latitude",
    "lon_d": "Delivery_location_longitude",
}
for k, v in COLS.items():
    if v not in df.columns:
        raise KeyError(f"Missing column: {v}")

# -------------------------
# 1) Parameters
# -------------------------
USE_ROI   = True
CENTER_LAT = 12.959861
CENTER_LON = 77.579225
RADIUS_KM  = 50.0
MODE = "both"  # {"both","either","midpoint"}

# -------------------------
# 2) Utility Functions
# -------------------------
def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dφ = lat2 - lat1
    dλ = lon2 - lon1
    a = np.sin(dφ/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dλ/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

def norm_traffic(s: pd.Series) -> pd.Series:
    s = s.astype("string").str.strip().str.lower()
    mapping = {
        "low":"low", "low ":"low",
        "medium":"medium", "moderate":"medium",
        "high":"high",
        "jam":"jam", "traffic jam":"jam", "congestion":"jam",
        "nan":np.nan, "":np.nan, "none":np.nan, "unknown":np.nan
    }
    y = s.map(mapping).fillna(s)
    return y.where(y.isin(["low","medium","high","jam"]), other=np.nan)

def parse_time_30min_adjusted(series: pd.Series) -> pd.Series:
    """'HH:MM' or 'HH:MM:SS' → 30min intervals, 3AM adjustment, invalid → NaN"""
    def _one(x):
        if pd.isna(x):
            return np.nan
        x = str(x).strip()
        m = re.match(r"^\s*(\d{1,2}):(\d{2})(?::(\d{2}))?\s*$", x)
        if not m:
            return np.nan
        h = int(m.group(1))
        min_val = int(m.group(2))
        if not (0 <= h <= 23 and 0 <= min_val <= 59):
            return np.nan

        # Adjust hours < 3 to next day
        if h < 3:
            h += 24

        # Convert to 30-min intervals
        time_30min = h * 2 + (1 if min_val >= 30 else 0)
        return time_30min
    return series.apply(_one).astype("Float64")

def to_int_safe(df_or_s):
    """Replace inf→NaN, fill NaN→0, cast to int safely."""
    if isinstance(df_or_s, pd.DataFrame):
        return (df_or_s.replace([np.inf, -np.inf], np.nan).fillna(0).round(0).astype(int))
    return (df_or_s.replace([np.inf, -np.inf], np.nan).fillna(0).round(0).astype(int))

def time_30min_to_display_label(time_30min):
    """Convert 30min index to display label (e.g., 48→00:00, 49→00:30, 50→01:00)"""
    hour = time_30min // 2
    min_part = "00" if time_30min % 2 == 0 else "30"

    # Convert back to 24-hour format for display
    if hour >= 24:
        hour = hour - 24

    return f"{hour:02d}:{min_part}"

# -------------------------
# 3) Time Parsing
# -------------------------
time_30min = parse_time_30min_adjusted(df[COLS["t_ordered"]])
traffic = norm_traffic(df[COLS["traffic"]])

# -------------------------
# 4) ROI Mask
# -------------------------
if USE_ROI:
    lat_r = pd.to_numeric(df[COLS["lat_r"]], errors="coerce").to_numpy()
    lon_r = pd.to_numeric(df[COLS["lon_r"]], errors="coerce").to_numpy()
    lat_d = pd.to_numeric(df[COLS["lat_d"]], errors="coerce").to_numpy()
    lon_d = pd.to_numeric(df[COLS["lon_d"]], errors="coerce").to_numpy()
    if CENTER_LAT is None:
        CENTER_LAT = float(np.nanmean(np.concatenate([lat_r, lat_d])))

    d_center_rest = haversine_km(lat_r, lon_r, CENTER_LAT, CENTER_LON)
    d_center_dest = haversine_km(lat_d, lon_d, CENTER_LAT, CENTER_LON)
    m_rest = np.isfinite(d_center_rest) & (d_center_rest <= RADIUS_KM)
    m_dest = np.isfinite(d_center_dest) & (d_center_dest <= RADIUS_KM)

    if   MODE == "both":     roi_mask = m_rest & m_dest
    elif MODE == "either":   roi_mask = m_rest | m_dest
    elif MODE == "midpoint":
        lat_mid = (lat_r + lat_d) / 2.0
        lon_mid = (lon_r + lon_d) / 2.0
        d_mid   = haversine_km(lat_mid, lon_mid, CENTER_LAT, CENTER_LON)
        roi_mask = np.isfinite(d_mid) & (d_mid <= RADIUS_KM)
    else:
        raise ValueError("MODE must be one of {'both','either','midpoint'}")
else:
    roi_mask = np.ones(len(df), dtype=bool)

# -------------------------
# 5) Final Filter & Analysis DataFrame
# -------------------------
valid = roi_mask & traffic.notna().to_numpy() & time_30min.notna().to_numpy()
data = pd.DataFrame({
    "time_30min": time_30min[valid].astype(int),
    "traffic": traffic[valid].to_numpy()
})

print(f"[INFO] total rows={len(df):,}, valid rows={len(data):,} "
      f"({(len(data)/len(df)*100 if len(df)>0 else 0):.2f}%) | ROI={'ON' if USE_ROI else 'OFF'}")

if data.empty:
    print("No valid data for time-based traffic analysis.")
else:
    levels = ["low","medium","high","jam"]

    # 30min interval counts
    time_range = sorted(data["time_30min"].unique())
    counts = (
        data.groupby(["time_30min","traffic"]).size()
            .unstack("traffic")
            .reindex(index=time_range, columns=levels)
    )
    counts = to_int_safe(counts)

    # Shares (%)
    denom = counts.sum(axis=1).replace(0, np.nan)
    shares = (counts.div(denom, axis=0) * 100).fillna(0)

    # ---- Table Output
    print("\n[30-min interval traffic counts]")
    counts_display = counts.copy()
    counts_display.index = [time_30min_to_display_label(t) for t in counts_display.index]
    print(counts_display.to_string())

    print("\n[30-min interval traffic shares (%)]")
    shares_display = shares.copy()
    shares_display.index = [time_30min_to_display_label(t) for t in shares_display.index]
    print(shares_display.round(2).to_string())

    # ---- 1) Stacked Bar Chart (shares, %)
    plt.figure(figsize=(15,6))
    bottom = np.zeros(len(time_range))
    x_pos = np.arange(len(time_range))
    for lvl in levels:
        plt.bar(x_pos, shares[lvl].to_numpy(), bottom=bottom, label=lvl.title())
        bottom += shares[lvl].to_numpy()
    plt.title("Road Traffic Share by 30-min Intervals (%)")
    plt.xlabel("Time (30-min intervals)")
    plt.ylabel("Share (%)")
    # Show every 2 hours (4 intervals)
    xtick_positions = range(0, len(time_range), 4)
    xtick_labels = [time_30min_to_display_label(time_range[i]) for i in xtick_positions]
    plt.xticks(xtick_positions, xtick_labels, rotation=45)
    plt.grid(True, axis='y', linewidth=0.5, alpha=0.5)
    plt.legend()
    plt.tight_layout()

    # ---- 2) Line Chart (shares, %)
    plt.figure(figsize=(15,6))
    for lvl in levels:
        plt.plot(x_pos, shares[lvl].to_numpy(), linewidth=2, marker='o', label=lvl.title())
    plt.title("Road Traffic Share by 30-min Intervals (%) – Line")
    plt.xlabel("Time (30-min intervals)")
    plt.ylabel("Share (%)")
    plt.xticks(xtick_positions, xtick_labels, rotation=45)
    plt.grid(True, linewidth=0.5, alpha=0.5)
    plt.legend()
    plt.tight_layout()

    # ---- 3) Heatmap (counts)
    plt.figure(figsize=(12,6))
    H = counts.to_numpy().T  # shape=(4, len(time_range))
    plt.imshow(H, aspect="auto", origin="lower",
               extent=[-0.5, len(time_range)-0.5, -0.5, len(levels)-0.5])
    plt.yticks(ticks=range(len(levels)), labels=[lvl.title() for lvl in levels])
    plt.xticks(ticks=xtick_positions, labels=xtick_labels, rotation=45)
    plt.colorbar(label="Count")
    plt.title("Heatmap: Count by 30-min Interval × Traffic")
    plt.xlabel("Time (30-min intervals)")
    plt.ylabel("Traffic level")
    plt.tight_layout()

    plt.show()
```

```python
# === Delivery Distance Analysis by 30-min Intervals with 3AM Adjustment ===
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# -------------------------
# 0) Data Loading
# -------------------------
try:
    assert isinstance(df, pd.DataFrame)
except NameError:
    CSV_PATH = "example/train.csv"
    df = pd.read_csv(CSV_PATH)

COLS = {
    "lat_r": "Restaurant_latitude",
    "lon_r": "Restaurant_longitude",
    "lat_d": "Delivery_location_latitude",
    "lon_d": "Delivery_location_longitude",
    "t_ordered": "Time_Orderd",
}
for k, v in COLS.items():
    if v not in df.columns:
        raise KeyError(f"Missing column: {v}")

# -------------------------
# 1) Parameters
# -------------------------
USE_ROI    = True
CENTER_LAT = 12.959861
CENTER_LON = 77.579225
RADIUS_KM  = 50.0
MODE       = "both"  # {"both","either","midpoint"}

MIN_DISTANCE = 0.0
BIN_W_DIST  = 5.0

# -------------------------
# 2) Utility Functions
# -------------------------
def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dφ = lat2 - lat1
    dλ = lon2 - lon1
    a = np.sin(dφ/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dλ/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

def parse_time_30min_adjusted(series: pd.Series) -> pd.Series:
    """'HH:MM' or 'HH:MM:SS' → 30min intervals with 3AM adjustment, invalid → NaN"""
    def _one(x):
        if pd.isna(x):
            return np.nan
        x = str(x).strip()
        m = re.match(r"^\s*(\d{1,2}):(\d{2})(?::(\d{2}))?\s*$", x)
        if not m:
            return np.nan
        h = int(m.group(1))
        min_val = int(m.group(2))
        if not (0 <= h <= 23 and 0 <= min_val <= 59):
            return np.nan

        # Adjust hours < 3 to next day
        if h < 3:
            h += 24

        # Convert to 30-min intervals
        time_30min = h * 2 + (1 if min_val >= 30 else 0)
        return time_30min
    return series.apply(_one).astype("Float64")

def to_int_safe(df_or_s):
    if isinstance(df_or_s, pd.DataFrame):
        return (df_or_s.replace([np.inf, -np.inf], np.nan).fillna(0).round(0).astype(int))
    return (df_or_s.replace([np.inf, -np.inf], np.nan).fillna(0).round(0).astype(int))

def time_30min_to_display_label(time_30min):
    """Convert 30min index to display label"""
    hour = time_30min // 2
    min_part = "00" if time_30min % 2 == 0 else "30"

    # Convert back to 24-hour format for display
    if hour >= 24:
        hour = hour - 24

    return f"{hour:02d}:{min_part}"

# -------------------------
# 3) Coordinate/Time Parsing
# -------------------------
lat_r = pd.to_numeric(df[COLS["lat_r"]], errors="coerce").to_numpy()
lon_r = pd.to_numeric(df[COLS["lon_r"]], errors="coerce").to_numpy()
lat_d = pd.to_numeric(df[COLS["lat_d"]], errors="coerce").to_numpy()
lon_d = pd.to_numeric(df[COLS["lon_d"]], errors="coerce").to_numpy()

time_30min = parse_time_30min_adjusted(df[COLS["t_ordered"]])

# -------------------------
# 4) ROI Mask
# -------------------------
if USE_ROI:
    if CENTER_LAT is None:
        CENTER_LAT = float(np.nanmean(np.concatenate([lat_r, lat_d])))
    d_center_rest = haversine_km(lat_r, lon_r, CENTER_LAT, CENTER_LON)
    d_center_dest = haversine_km(lat_d, lon_d, CENTER_LAT, CENTER_LON)
    m_rest = np.isfinite(d_center_rest) & (d_center_rest <= RADIUS_KM)
    m_dest = np.isfinite(d_center_dest) & (d_center_dest <= RADIUS_KM)
    if   MODE == "both":     roi_mask = m_rest & m_dest
    elif MODE == "either":   roi_mask = m_rest | m_dest
    elif MODE == "midpoint":
        lat_mid = (lat_r + lat_d) / 2.0
        lon_mid = (lon_r + lon_d) / 2.0
        d_mid   = haversine_km(lat_mid, lon_mid, CENTER_LAT, CENTER_LON)
        roi_mask = np.isfinite(d_mid) & (d_mid <= RADIUS_KM)
    else:
        raise ValueError("MODE must be one of {'both','either','midpoint'}")
else:
    roi_mask = np.ones(len(df), dtype=bool)

# -------------------------
# 5) Distance Calculation + Final Filter
# -------------------------
dist_km = haversine_km(lat_r, lon_r, lat_d, lon_d)
valid_dist = np.isfinite(dist_km) & (dist_km > MIN_DISTANCE)
valid = roi_mask & valid_dist & time_30min.notna().to_numpy()

data = pd.DataFrame({
    "distance_km": dist_km[valid],
    "time_30min": time_30min[valid].astype(int)
})

print(f"[INFO] total rows={len(df):,}, valid rows={len(data):,} "
      f"({(len(data)/len(df)*100 if len(df)>0 else 0):.2f}%) | ROI={'ON' if USE_ROI else 'OFF'}")

if data.empty:
    print("No valid data for time-of-day distance analysis.")
else:
    # -------------------------
    # 6) 30-min Interval Summary
    # -------------------------
    def p25(s): return np.percentile(s, 25)
    def p75(s): return np.percentile(s, 75)

    time_range = sorted(data["time_30min"].unique())
    summary = (
        data.groupby("time_30min")["distance_km"]
            .agg(count="count",
                 mean_km="mean",
                 median_km="median",
                 p25_km=p25,
                 p75_km=p75,
                 std_km=lambda s: s.std(ddof=1) if s.size>1 else 0.0)
            .reindex(time_range)
    )

    counts_only = to_int_safe(summary[["count"]])
    print("\n[30-minute interval summary of delivery distance]")
    to_print = summary.copy()
    to_print["count"] = counts_only["count"]
    to_print.index = [time_30min_to_display_label(i) for i in to_print.index]
    print(to_print.round(3).to_string())

    # -------------------------
    # 7) Graph 1: Median + IQR Band
    # -------------------------
    idx = summary.index.to_numpy()
    med = summary["median_km"].to_numpy()
    p25v = summary["p25_km"].to_numpy()
    p75v = summary["p75_km"].to_numpy()

    plt.figure(figsize=(15,6))
    x_pos = np.arange(len(time_range))
    plt.plot(x_pos, med, linewidth=2)
    plt.fill_between(x_pos, p25v, p75v, alpha=0.25)
    plt.title("Median Delivery Distance by 30-min Intervals (with IQR)")
    plt.xlabel("Time (30-min intervals)")
    plt.ylabel("Distance (km)")

    # Show every 2 hours (4 intervals)
    xtick_positions = range(0, len(time_range), 4)
    xtick_labels = [time_30min_to_display_label(time_range[i]) for i in xtick_positions]
    plt.xticks(xtick_positions, xtick_labels, rotation=45)
    plt.grid(True, linewidth=0.5, alpha=0.5)
    plt.tight_layout()

    # -------------------------
    # 8) Graph 2: Order Count
    # -------------------------
    plt.figure(figsize=(15,5))
    plt.bar(x_pos, counts_only["count"].to_numpy())
    plt.title("Valid Orders per 30-min Interval")
    plt.xlabel("Time (30-min intervals)")
    plt.ylabel("Count")
    plt.xticks(xtick_positions, xtick_labels, rotation=45)
    plt.grid(True, axis='y', linewidth=0.5, alpha=0.5)
    plt.tight_layout()

    # -------------------------
    # 9) Graph 3: 2D Heatmap
    # -------------------------
    upper_d = BIN_W_DIST * np.ceil(max(1.0, data["distance_km"].max()) / BIN_W_DIST)
    d_edges = np.arange(0.0, upper_d + BIN_W_DIST + 1e-9, BIN_W_DIST)
    t_edges = np.arange(-0.5, len(time_range)-0.5 + 1.0, 1.0)

    H, xedges, yedges = np.histogram2d(
        [time_range.index(t) for t in data["time_30min"].to_numpy()],  # Convert to index position
        data["distance_km"].to_numpy(),
        bins=[t_edges, d_edges]
    )

    plt.figure(figsize=(12,6))
    plt.imshow(H.T, origin="lower", aspect="auto",
               extent=[t_edges[0], t_edges[-1], d_edges[0], d_edges[-1]])
    plt.colorbar(label="Count")
    plt.title("Heatmap: Count by 30-min Interval × Distance")
    plt.xlabel("Time (30-min intervals)")
    plt.ylabel("Distance (km)")
    plt.xticks(xtick_positions, xtick_labels, rotation=45)
    plt.tight_layout()

    plt.show()
```
